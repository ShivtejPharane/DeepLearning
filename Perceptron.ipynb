```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("student_placement_dataset.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x=df['CGPA'], y=df['ResumeScore'], hue=df['Placed'])
```

```python
x = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron(random_state=1)
p.fit(x,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.values, y.values, clf=p)
```
